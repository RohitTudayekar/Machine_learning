```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/Social_Network_Ads (1).csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
#dropping the USer ID Column
df=data.drop(['User ID'],axis=1)
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].apply({'Male':0,'Female':1}.get)
```

```python
df.head()
```

```python
x=df[['Gender','Age','EstimatedSalary']] #independent Columns
y=df['Purchased']#dependent Columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2)
```

```python
#importing KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
```

```python
predictions=knn.predict(x_test)
```

```python
#accuracy
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,predictions)
ac=accuracy_score(y_test,predictions)
```
