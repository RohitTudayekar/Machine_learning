```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#step 2 #data processing
#convert categorical variables od dataset into numerical variables bu using ONE HOT ENCODING Techniques
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
```

```python
data.head()
```

```python
data['sex']=data['sex'].apply({'male':1,'female':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':0,'northeast':2,'northwest':3}.get)
```

```python
data.head()
```

```python
#dividing the dataset into dependant variables and independent variables colunns
X=data[['age','sex','bmi','children','smoker','region']] # X contains independent variables
y=data['charges']    # y contain dependent variable
```

```python
X.head()
```

```python
y.head()
```

```python
# splitting dataset into training and testing test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2) #20% of the dataset will be used for testing & 80% of data willbe used for testing
```

```python
#creating Machine Learning model
from sklearn.linear_model import LinearRegression
linearmodel=LinearRegression()
linearmodel.fit(X_train,y_train) # training is taking place
```

```python
#predictions
linearmodel.predict(X_test)
```

```python
# cal.accuracy of model
linearmodel.score(X,y)
```

```python
# predicting w.r.t. new data
data_new={'age':62,'sex':0,'bmi':26,'children':2,'smoker':1,'region':3}
index=[1] # serial number
my_data=pd.DataFrame(data_new,index)
print(my_data)
```

```python
#Charges Allocated
my_data_charges=linearmodel.predict(my_data)
print("Charges allocated to new customer is",my_data_charges)
```
